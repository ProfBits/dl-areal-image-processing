```python
from mmseg.apis import inference_model, init_model, show_result_pyplot
import mmcv
import cv2
import matplotlib.pyplot as plt
import os
```

```python
img = 'demopics/32691_5335_masked.tif'  # or img = mmcv.imread(img), which will only load it once
# img = mmcv.imread(img)
img = cv2.imread(img)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img = cv2.resize(img, (512,512))

plt.figure(figsize=(10, 10))
plt.imshow(img)
plt.axis('off')
plt.show()
```

# DeepLabV3+

```python
config_file = r'C:\Users\henha\DeepLearning\mmsegmentation\configs\deeplabv3plus\deeplabv3plus_r18-d8_4xb4-80k_potsdam-512x512.py'
checkpoint_file = 'checkpoints/deeplabv3plus_r18-d8_512x512_80k_potsdam_20211219_020601-75fd5bc3.pth'

model = init_model(config_file, checkpoint_file)

# test a single image and show the results
result = inference_model(model, img)
# visualize the results in a new window
res = show_result_pyplot(model, img, result, show=True, draw_pred=True)
plt.imshow(res)
```
