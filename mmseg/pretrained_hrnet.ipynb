# HRNet

```python
from mmseg.apis import inference_model, init_model, show_result_pyplot
import mmcv
import cv2
import matplotlib.pyplot as plt
```

```python
img_raw = 'demopics/32686_5337.tif'
img_masked = 'demopics/32686_5337_masked.tif'  # or img = mmcv.imread(img), which will only load it once

img_raw = cv2.imread(img_raw)
img_raw = cv2.cvtColor(img_raw, cv2.COLOR_BGR2RGB)
# img_raw = cv2.resize(img_raw, (512,512))

img_masked = cv2.imread(img_masked)
img_masked = cv2.cvtColor(img_masked, cv2.COLOR_BGR2RGB)
# img_masked = cv2.resize(img_masked, (512,512))

# plt.imshow(img_raw)
# plt.imshow(img_masked)
```

```python
img_raw_crop = img_raw[0:512, 0:512]
img_masked_crop = img_masked[0:512, 0:512]

# plt.imshow(img_raw_crop)
# plt.imshow(img_masked_crop)
```

## LoveDA

```python
loveDa_r18_config = r"mmsegmentation\configs\hrnet\fcn_hr18_4xb4-80k_loveda-512x512.py"
loveDa_r18_checkpoint = r"checkpoints\HRNET\fcn_hr18_512x512_80k_loveda_20211210_203952-93d9c3b3.pth"

loveDa_r18_model = init_model(loveDa_r18_config, loveDa_r18_checkpoint)
```

```python
# test a single image and show the results
loveDa_r18_im_raw = inference_model(loveDa_r18_model, img_raw)
# visualize the results in a new window
loveDa_r18_raw_result = show_result_pyplot(loveDa_r18_model, img_raw, loveDa_r18_im_raw, show=True, draw_pred=True, withLabels=False)
plt.imshow(loveDa_r18_raw_result)
```

```python
# test a single image and show the results
loveDa_r18_im_raw_crop = inference_model(loveDa_r18_model, img_raw_crop)
# visualize the results in a new window
loveDa_r18_raw_crop_result = show_result_pyplot(loveDa_r18_model, img_raw_crop, loveDa_r18_im_raw_crop, show=True, draw_pred=True, withLabels=False)
plt.imshow(loveDa_r18_raw_crop_result)
```
