```python
from mmseg.apis import inference_model, init_model, show_result_pyplot
import mmcv
import cv2
import matplotlib.pyplot as plt
import os
```

```python
img = 'demopics/32691_5335_masked.tif'  # or img = mmcv.imread(img), which will only load it once
img_raw = 'demopics/32691_5335.tif'
# img = mmcv.imread(img)
img = cv2.imread(img)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# img = cv2.resize(img, (512,512))

img_raw = cv2.imread(img_raw)
img_raw = cv2.cvtColor(img_raw, cv2.COLOR_BGR2RGB)
# img_raw = cv2.resize(img_raw, (512,512))

# plt.figure(figsize=(10, 10))
# plt.imshow(img)
# plt.axis('off')
# plt.show()
```

```python
config_file = r"mmsegmentation\configs\pspnet\pspnet_r18-d8_4xb4-80k_loveda-512x512.py"
checkpoint_file = r"checkpoints\PSPNET\pspnet_r18-d8_512x512_80k_loveda_20211105_052100-b97697f1.pth"

model = init_model(config_file, checkpoint_file)
```

```python
# test a single image and show the results
result_crop = inference_model(model, crop_img_raw)
# visualize the results in a new window
res_crop = show_result_pyplot(model, crop_img_raw, result_crop, show=True, draw_pred=True)
plt.imshow(res_crop)
```

```python
# test a single image and show the results
result = inference_model(model, img)
# visualize the results in a new window
res = show_result_pyplot(model, img, result, show=True, draw_pred=True)
plt.imshow(res)
```

```python
# test a single image and show the results
result_raw = inference_model(model, img_raw)
# visualize the results in a new window
res_raw = show_result_pyplot(model, img_raw, result_raw, show=True, draw_pred=True)
plt.imshow(res_raw)
```

```python
config_file = r"mmsegmentation\configs\pspnet\pspnet_r101-d8_4xb4-80k_loveda-512x512.py"
checkpoint_file = r"checkpoints\PSPNET\pspnet_r101-d8_512x512_80k_loveda_20211104_153212-1c06c6a8.pth"

model = init_model(config_file, checkpoint_file)

# test a single image and show the results
result = inference_model(model, img)
result_raw = inference_model(model, img_raw)
# visualize the results in a new window
figure, axis = plt.subplots(1, 2)
result = show_result_pyplot(model, img, result, show=True, draw_pred=True)
axis[0].imshow(result)
axis[0].set_title('RESULT')

result_raw = show_result_pyplot(model, img_raw, result_raw, show=True, draw_pred=True)
axis[1].imshow(result_raw)
axis[1].set_title('RAW')
```
